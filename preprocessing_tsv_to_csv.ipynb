```python
data2 = data.iloc[:,:].values
```

```python
def ind_type(char):
    if char == "IND_TOTAL":
        return "All individuals"
    elif char == "Y16_24":
        return "Individuals, 16 TO 24 years old"
    elif char == "Y25_54":
        return "Individuals, 25 TO 54 years old"
    elif char == "Y55_74":
        return "Individuals, 55 TO 74 years old"
    elif char == "M_Y16_74":
        return "Males, 16 TO 74 years old"
    elif char == "F_Y16_74":
        return "Females, 16 TO 74 years old"
    elif char == "I0_2":
        return "Individuals with no or low formal education"
    elif char == "I3_4":
        return "individuals with medium formal education"
    elif char == "I5_8":
        return "Individuals with high formal education"
    elif char == "EMPL_UNE":
        return "Active labour force (employed and unemployed)"
    elif char == "RETIR_OTHER":
        return "Retired and other inactive"
    elif char == "SAL_SELF_FAM":
        return "Employees, self-employed, familly workers"
    elif char == "STUD":
        return "students"
    elif char == "UNE":
        return "Unemployed"
```

```python
for i in data2:
    i[1] = ind_type(i[1])
    for j in range(4,len(i)):
        if ":" in str(i[j]):
            i[j]="NA"
        elif len(str(i[j]))>=2:
            i[j]=str(i[j])[0:2]
```
